```python
# Priority-weighted Load Restoration Problem for Improved Resilience in Distribution Systems
# MPC 
# The data analytics (pre-processing and post-processing) is written in Python
# While the MPC is written in Julia and called from Python
```

```python
# Import required Python packages/modules

import pandas as pd 
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sb
sb.set()
```

```python
# Calling Julia Packages from Python
# Install Julia packages required by PyJulia

import julia
from julia.api import Julia
jl = Julia(compiled_modules=False)       # Turn off compilation cache
jl = julia.Julia()
```

```python
# Import the Julia modules required  

from julia import Base
from julia import Main
from julia import JuMP
from julia import GLPK
from julia import Ipopt
from julia import Cbc
```

```python
# Load demand

load_new = [33, 34, 8.5, 85, 60, 60, 58, 115, 64, 85]
importance_factor = [1.0, 1.0, 0.9, 0.85, 0.8, 0.65, 0.45, 0.4, 0.3, 0.3]
num_time_steps = 72

# Create time series of the load demand - constant time series

Load = list()
for i in range(len(load_new)):
        Load.append([load_new[i]] * num_time_steps)
Load = np.array(Load)

time = range(num_time_steps)
time = np.array(time)
for i in range(len(Load)):
    plt.plot(time, Load[i,:],linewidth = 3)
    plt.xlabel('Time (hr)', fontsize=15)
    plt.ylabel('Power (kW)', fontsize=15)
    plt.tick_params(labelsize=12)
    plt.xticks(range(0, 73, 12), [str(int(c/12)) + ":00" for c in range(0, 73, 12)])
    plt.rcParams["font.family"] = "Times New Roman"
    #plt.show()
```

```python
# Load exogenous data - load profile & renewable energy generation profile
    # Normalized (between 0 and 1) time series data with 5 minutes resolution

exo_data_path = "/Users/aeseye/Desktop/AGM-Resilience-Project/agm_resilience/agm_resiliency/dss_gym_env/dss_envs/envs/data/13BusCustomed/five_min_load_profile.csv"
exo_data = pd.read_csv(exo_data_path)

# Extract the RE generations

Pwind = exo_data['wind_gen']
Ppvv = exo_data['pv_gen']

# Define the rated capacities (kW) of the renewable generators

WindMAX = 150
pvMAX = 300

time = range(len(Pwind))
time = np.array(time)
plt.plot(time, WindMAX*Pwind, label='Wind', linewidth = 3)
plt.plot(time, pvMAX*Ppvv, label='PV', linewidth = 3)
plt.xlabel('Time [5-min]', fontsize=15)
plt.ylabel('Power [kW]', fontsize=15)
plt.tick_params(labelsize=12)
plt.rcParams["font.family"] = "Times New Roman"
plt.legend()
plt.show()
```

```python
# Pick 6 hours (i.e., 72 5-minutes) length exo data
    # for the 6h horizon (look-ahead) priority-weighted load pick-up optimization

DURATION = 72
START_IDX = 1217
NUM_OF_LOAD_BUS = 10
FIVE_MIN_TO_HOUR = 5/60
Δt = FIVE_MIN_TO_HOUR

# Obtain the actual power (kW) time series of the renewable generations
load_profile = Load
wind_profile = WindMAX*Pwind[START_IDX: START_IDX+DURATION]
pv_profile = pvMAX*Ppvv[START_IDX: START_IDX+DURATION]
```

```python
# Actual load demand (real power) at each time step

Actual_Load_Real_t = list()
for ix in range(num_time_steps):
    Load_t = sum(load_profile[:,ix])
    Actual_Load_Real_t.append(Load_t)
Actual_Load_Real_t = np.array(Actual_Load_Real_t)

# Actual energy demand at each node -- in kWh  
    # (Total energy at a node makes sense for time series loads)

Energy_n = list()
for jx in range(len(load_profile)):
    ener_n = sum(load_profile[jx,:])*Δt
    Energy_n.append(ener_n)
Energy_n = np.array(Energy_n)
```

```python
# Define (other) constant parameters

num_nodes = NUM_OF_LOAD_BUS
Pl_act = load_profile
Pwt = wind_profile
Ppv = pv_profile
load_priority_weight = importance_factor
PmtMIN = 0
PmtMAX = 300
PchMAX = 200
PdischMAX = 200
Emt = 1000         # kWh
SOCMIN = 0.20      # %
SOCMAX = 1.0 
SOC0 = 0.90
eff_ch = 0.95
eff_disch = 0.90
Ces = 800          #kWh
alpha =  0.02      #$/kWh -- penalty for wind power curtailment 
beta = 0.02        #$/kWh -- penalty for PV power curtailment
```

```python
# Plot the Load Priority Weight

node_idx = range(NUM_OF_LOAD_BUS)
node_idx = np.array(node_idx)
plt.plot(node_idx, load_priority_weight, linewidth = 3)
plt.xlabel('Node', fontsize=15)
plt.ylabel('Load Priority Weight', fontsize=15)
plt.tick_params(labelsize=12)
plt.rcParams["font.family"] = "Times New Roman"
plt.show()
```

```python
# Hereafter call the controller written in Julia-JuMP
```

```python
# Call/Run the MPC (from Julia)

SOC_level = SOC0
Emt_level = Emt

P_restoredd = np.empty((0, num_nodes))
Pmt_genn = ([])
Pbatt = ([])
SOCbatt = ([])
Pwt_cutt = ([])
Ppv_cutt = ([])
 
for k in range(1,num_time_steps+1):
    jl.include("MPC")
    P_restored, Pmt_gen, Pbat, SOCbat, Pwt_cut, Ppv_cut = jl.run_mpc(SOC_level, Emt_level, k)
    P_restoredd = np.append(P_restoredd, [P_restored], axis=0)
    Pmt_genn = np.append(Pmt_genn, Pmt_gen)
    Pbatt = np.append(Pbatt, Pbat)
    SOCbatt = np.append(SOCbatt, SOCbat)
    Pwt_cutt = np.append(Pwt_cutt, Pwt_cut)
    Ppv_cutt = np.append(Ppv_cutt, Ppv_cut)
    SOC_level = SOCbat
    Emt_level = Emt_level - Pmt_gen*Δt
P_restoredd = np.transpose(P_restoredd)
```

```python
# Restored load at each time step

P_restored_t = list()
for jdx in range(num_time_steps):
    P_restored_t.append(sum(P_restoredd[:,jdx]))
P_restored_t = np.array(P_restored_t)

# Restored energy at each node

E_restored_n = list()
for qdx in range(num_nodes):
    E_restored_n.append((sum(P_restoredd[qdx,:]))*Δt)
E_restored_n = np.array(E_restored_n)
```

```python
# Plot the results - restored load and dispatch of the controllable resources

t = range(num_time_steps)
t = np.array(t)
plt.plot(t, P_restored_t, label='Restored Load', linewidth = 3)
plt.plot(t, Pwt, label='Wind', linewidth = 3)
plt.plot(t, Ppv, label='PV', linewidth = 3)
plt.plot(t, Pmt_genn, label='Microturbine', linewidth = 3)
plt.plot(t, Pbatt, label='Battery', linewidth = 3, color = 'blueviolet')
plt.xlabel('Time [h]', fontsize=15)
plt.ylabel('Power [kW]', fontsize=15)
plt.tick_params(labelsize=12)
plt.xticks(range(0, 73, 12), [str(int(c/12)) + ":00" for c in range(0, 73, 12)])
plt.rcParams["font.family"] = "Times New Roman"
#plt.grid()
plt.legend()
plt.show()
```

```python
# Plot the renewable power curtailment

t = range(num_time_steps)
t = np.array(t)
plt.plot(t, Pwt_cutt, label='Curtailed Wind', linewidth = 3)
plt.plot(t, Ppv_cutt, label='Curtailed PV', linewidth = 3)
plt.xlabel('Time [h]', fontsize=15)
plt.ylabel('Power [kW]', fontsize=15)
plt.tick_params(labelsize=12)
plt.xticks(range(0, 73, 12), [str(int(c/12)) + ":00" for c in range(0, 73, 12)])
plt.rcParams["font.family"] = "Times New Roman"
#plt.grid()
plt.legend()
plt.show()
```

```python
# Change in Energy of Battery and Microturbine

Emt_gen = list()
Emt_gen.append(Emt - Pmt_genn[0]*Δt)
for v in range(1,num_time_steps):
    Emt_gen.append(Emt_gen[v-1] - Pmt_genn[v]*Δt)
Emt_gen = np.array(Emt_gen)

t = range(num_time_steps)
t = np.array(t)
plt.plot(t, Emt_gen, label='Microturbine', linewidth = 3)
plt.plot(t, Ces*SOCbatt, label='Battery', linewidth = 3)
plt.xlabel('Time [h]', fontsize=15)
plt.ylabel('Energy [kWh]', fontsize=15)
plt.tick_params(labelsize=12)
plt.xticks(range(0, 73, 12), [str(int(c/12)) + ":00" for c in range(0, 73, 12)])
plt.rcParams["font.family"] = "Times New Roman"
#plt.grid()
plt.legend()
plt.show()
```

```python
# Plot the battery state of charge (SOC) - %

t = range(num_time_steps)
t = np.array(t)
plt.bar(t,  100*SOCbatt, width = 0.8, color = ['red'])
plt.xlabel('Time [h]', fontsize=15)
plt.ylabel('SOC [%]', fontsize=15)
plt.tick_params(labelsize=12)
plt.xticks(range(0, 73, 12), [str(int(c/12)) + ":00" for c in range(0, 73, 12)])
plt.rcParams["font.family"] = "Times New Roman"
#plt.grid()
plt.show()
```

```python
# Plot the percentage of the restored load per time step
    # %Restored Load = Restored Load/Actual Load Demand
    
t = range(num_time_steps)
t = np.array(t)
P_restored_t_perc = (P_restored_t / Actual_Load_Real_t) *100
plt.plot(t, P_restored_t_perc, linewidth = 3)
plt.xlabel('Time [h]', fontsize=15)
plt.ylabel('Load Served [%]', fontsize=15)
plt.tick_params(labelsize=12)
plt.xticks(range(0, 73, 12), [str(int(c/12)) + ":00" for c in range(0, 73, 12)])
plt.rcParams["font.family"] = "Times New Roman"
#plt.grid()
plt.show()
```

```python
# Plot the percentage of the restored energy per node

n = range(num_nodes)
n = np.array(n)
E_restored_n_perc = (E_restored_n / Energy_n) *100
plt.plot(n, E_restored_n_perc, linewidth = 3)
plt.xlabel('Node', fontsize=15)
plt.ylabel('Energy Supplied [%]', fontsize=15)
plt.tick_params(labelsize=12)
plt.xticks(range(0, 10, 1), [str(int(c)) for c in range(1, 11, 1)])
plt.rcParams["font.family"] = "Times New Roman"
#plt.grid()
plt.show()
```
